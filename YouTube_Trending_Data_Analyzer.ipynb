```python
import pandas as pd
df = pd.read_csv(r"D:\Python Codes\USvideos.csv")

# Basic info
print("Dataset shape:", df.shape)     #Dataset shape: (40949, 16)
print("\nColumns:", df.columns)       #Columns: Index(['video_id', 'trending_date', 'title', 'channel_title', 'category_id',
                                      #'publish_time', 'tags', 'views', 'likes', 'dislikes', 'comment_count',
                                      #'thumbnail_link', 'comments_disabled', 'ratings_disabled',
                                      #'video_error_or_removed', 'description'],
                                      

# Drop duplicates and nulls
df = df.drop_duplicates().dropna()

# Most viewed video
top_videos = df.nlargest(1, "views")[["title", "channel_title", "views"]]  #Top 10 Most Viewed Videos: title        channel_title  \
                                                                        #Childish Gambino - This Is America (Official V...  ChildishGambinoVEVO   

                                                                        #views  38547  225211923
print("\nTop 10 Most Viewed Videos:\n", top_videos)

# Average views per category
avg_views = df.groupby("category_id")["views"].mean().sort_values(ascending=False)   #category_id
                                                                                            #10    6.201003e+06
                                                                                            #1     3.106250e+06
                                                                                            #29    2.963884e+06
                                                                                            #20    2.620831e+06
                                                                                            #24    2.067883e+06
                                                                                            #17    2.025969e+06
                                                                                            #22    1.531835e+06
                                                                                            #23    1.480308e+06
                                                                                            #28    1.452627e+06
                                                                                            #2     1.355965e+06
                                                                                            #26    9.837301e+05
                                                                                            #43    9.035273e+05
                                                                                            #19    8.546196e+05
                                                                                            #15    8.311435e+05
                                                                                            #27    7.129408e+05
                                                                                            #25    5.925877e+05
print("\nAverage Views per Category:\n", avg_views)

# Most common category
most_common_cat = df["category_id"].mode()[0]        #Most Common Category ID: 24
print("\nMost Common Category ID:", most_common_cat)

# Top 10 channels with most trending videos
top_channels = df["channel_title"].value_counts().head(10)                                   #Top 10 Channels by Trending Videos:
                                                                                             #channel_title
                                                                                                 #ESPN                                      203
                                                                                                 #The Tonight Show Starring Jimmy Fallon    197
                                                                                                 #Vox                                       193
                                                                                                 #Netflix                                   193
                                                                                                 #TheEllenShow                              193
                                                                                                 #The Late Show with Stephen Colbert        187
                                                                                                 #Jimmy Kimmel Live                         186
                                                                                                 #Late Night with Seth Meyers               183
                                                                                                 #Screen Junkies                            182
                                                                                                 #NBA                                       181
print("\nTop 10 Channels by Trending Videos:\n", top_channels)
```
